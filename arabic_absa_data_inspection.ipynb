# Arabic ABSA - Data Loading & Quick Exploration
Simple notebook code to load Excel files and inspect key columns.

```python
import pandas as pd

# If files are in the same folder as this notebook/script, these names are enough.
# You can also use full paths (example shown for train file).
train_path = r"C:\\Users\\samam\\Desktop\\hack\\hackthon\\DeepX_train.xlsx"
valid_path = "DeepX_validation.xlsx"
unlabeled_path = "DeepX_unlabeled.xlsx"

# 1) Load all three files
train_df = pd.read_excel(train_path)
valid_df = pd.read_excel(valid_path)
unlabeled_df = pd.read_excel(unlabeled_path)

datasets = {
    "Train": train_df,
    "Validation": valid_df,
    "Unlabeled": unlabeled_df
}

for name, df in datasets.items():
    print("=" * 70)
    print(f"{name} Dataset")

    # 2) Shape
    print(f"Shape: {df.shape}")

    # 3) Column names
    print("Columns:", list(df.columns))

    # 4) First 5 rows
    print("\nFirst 5 rows:")
    display(df.head())

    # 5) Missing values
    print("Missing values per column:")
    print(df.isna().sum())

    # 6) Unique values in aspects (if available)
    if "aspects" in df.columns:
        print("\nUnique values in 'aspects':")
        print(df["aspects"].dropna().unique())

    # 7) Unique values in aspect_sentiments (if available)
    if "aspect_sentiments" in df.columns:
        print("\nUnique values in 'aspect_sentiments':")
        print(df["aspect_sentiments"].dropna().unique())

    # 8) Distribution of business_category and platform (if available)
    for col in ["business_category", "platform"]:
        if col in df.columns:
            print(f"\nDistribution of '{col}':")
            print(df[col].value_counts(dropna=False))

    print()
```
